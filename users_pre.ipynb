```python
prev = -1
use = 0
for i, row in users.iterrows():
    if row["user_id"] != prev:
        if (use % 100 == 0):
            print(use)
        if (use >= 1000):
            print(f"First {use} users done")
            break;
        use += 1
        userd.loc[(use-1), 'user_id'] = row['user_id']
        ind = userd.index[userd['user_id'] == row["user_id"]].tolist()[0]
        prev = row["user_id"]
    userd.at[ind, row["anime_id"]] = row["rating"]
```

```python
userd = userd.fillna(0)
userd.tail()
```

```python
user_ids = userd["user_id"]
user_ids.head()
```

```python
user_id2 = user_ids.apply(str)
user_id2.head()
```

```python
userd = userd.drop(["user_id"], axis=1)
userd.tail()
```

```python
cosine_sim = cosine_similarity(userd, userd)
print(cosine_sim)
```

```python
indices = pd.Series(user_id2)

def recommend(title, cosine_sim = cosine_sim):
    recommended_users = []
    idx = indices[indices == title].index[0]
    score_series = pd.Series(cosine_sim[idx]).sort_values(ascending = False)
    top_10_indices = list(score_series.iloc[1:11].index)
    
    for i in top_10_indices:
        recommended_users.append(list(user_id2)[i])
        
    return np.vstack((recommended_users,score_series.iloc[1:11].values)).T

recommend("2.0")
```
