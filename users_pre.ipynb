```python
userd = userd.fillna(0)
userd.tail()
```

```python
user_ids = userd["user_id"]
user_ids.head()
```

```python
user_id2 = user_ids.apply(str)
user_id2.head()
```

```python
userd = userd.drop(["user_id"], axis=1)
userd.tail()
```

```python
cosine_sim = cosine_similarity(userd, userd)
print(cosine_sim)
```

```python
indices = pd.Series(user_id2)

def recom(title, cosine_sim = cosine_sim):
    recommended_users = []
    idx = indices[indices == title].index[0]
    score_series = pd.Series(cosine_sim[idx]).sort_values(ascending = False)
    top_10_indices = list(score_series.iloc[1:11].index)
    
    for i in top_10_indices:
        recommended_users.append([list(user_id2)[i], i])
        
    return recommended_users

def recommend(search, cosine_sim=cosine_sim):
    top = recom(search, cosine_sim)[0]
    ind = indices[indices==search].index[0]
    ti = userd.loc[[ind, top[1]]]
    ti = ti.sort_values(by = top[1], axis = 1, ascending=False)
    exc = [column for column in ti.columns if ti.loc[top[1]][column] > 0]
    ti2 = ti.sort_values(by = ind, axis = 1, ascending=False)
    exc2 = [column for column in ti2.columns if ti2.loc[ind][column] > 0]
    return [x for x in exc2 if x not in exc][:10]

recommend("2.0")
```
