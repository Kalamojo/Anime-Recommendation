```python
import pandas as pd
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity

def to_1D(series):
    return pd.Series([x for _list in series for x in _list])
```

```python
url = "anime.csv"
anime = pd.read_csv(url)
anime.head()
```

```python
url = "anime2.csv"
anime_num = pd.read_csv(url)
anime_num = anime_num.drop(columns=["Unnamed: 0"], axis=1)
anime_num.head()
```

```python
url = "users2.csv"
users = pd.read_csv(url)
user_ids = users["user_id"].apply(str)
users = users.drop(["user_id"], axis=1)
users.head()
```

```python
cosine_simU = cosine_similarity(users, users)
print(cosine_simU)
```

```python
indicesU = pd.Series(user_ids)

def recom(title, cosine_sim):
    recommended_users = []
    idx = indicesU[indicesU == title].index[0]
    score_series = pd.Series(cosine_sim[idx]).sort_values(ascending = False)
    top_10_indices = list(score_series.iloc[1:11].index)
    
    for i in top_10_indices:
        recommended_users.append([list(user_ids)[i], i])
        
    return recommended_users

def recU(search, cosine_sim=cosine_simU):
    top = recom(search, cosine_sim)[0]
    ind = indicesU[indicesU==search].index[0]
    ti = users.loc[[ind, top[1]]]
    ti = ti.sort_values(by = top[1], axis = 1, ascending=False)
    exc = [column for column in ti.columns if ti.loc[top[1]][column] > 0]
    ti2 = ti.sort_values(by = ind, axis = 1, ascending=False)
    exc2 = [column for column in ti2.columns if ti2.loc[ind][column] > 0]
    return [x for x in exc2 if x not in exc][:10]

def recommendU(anime_ids):
    names = []
    for animes in anime_ids:
        names.append(anime[anime["MAL_ID"]==int(animes)]["Name"].values[0])
    return names

def recommendUser(user):
    ids = recU(user)
    return recommendU(ids)

recommendUser("2.0")
```

```python
cosine_simA = cosine_similarity(anime_num, anime_num)
print(cosine_simA)
```

```python
indicesA = pd.Series(anime["Name"])

def recommendAnime(title, cosine_sim = cosine_simA):
    recommended_anime = []
    idx = indicesA[indicesA == title].index[0]
    score_series = pd.Series(cosine_sim[idx]).sort_values(ascending = False)
    top_10_indices = list(score_series.iloc[1:11].index)
    
    for i in top_10_indices:
        recommended_anime.append(list(anime['Name'])[i])
        
    return recommended_anime

recommendAnime('Cowboy Bebop')
```
